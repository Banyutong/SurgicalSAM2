```python
from glob import glob
import os
import json
import re
```

```python
img_list = glob("/bd_byta6000i0/users/dataset/MedicalImage/CholecSeg8k/processed_scene/images/train/**/*.png", recursive=True)
```

```python
img_list
```

```python
file_name = os.path.split(img_list[0])[-1]
```

```python
str_match = re.match(r"(.*)(frame.*)",file_name)
```

```python
str_match.group(2)
```

```python
frames_info = []

for img_path in img_list:
    file_name = os.path.split(img_path)[-1]
    str_match = re.match(r"(.*)(frame.*)",file_name)
    video_id = str_match.group(1)
    path = img_path
    gt_path = os.path.splitext(img_path)[0] + "_color_mask.png"
    gt_path = gt_path.replace("images","labels")
    img_info = {
        "video_id": video_id,
        "file_path": path,
        "gt_path": gt_path,
        "file_name" : file_name
    }
    
    frames_info.append(img_info)
    
    
```

```python
class_dict = {
    0: "Black Background",
    1: "Abdominal Wall",
    2: "Liver",
    3: "Gastrointestinal Tract",
    4: "Fat",
    5: "Grasper",
    6: "Connective Tissue",
    7: "Blood",
    8: "Cystic Duct",
    9: "L-hook Electrocautery",
    10: "Gallbladder",
    11: "Hepatic Vein",
    12: "Liver Ligamen"
}
```

```python
raw_info = {
    "frames": frames_info,
    "class_dict": class_dict,
}
```

```python
with open("raw_info_cholec_seg8k.json", "w") as f:
    json.dump(raw_info, f, indent=4)
```
